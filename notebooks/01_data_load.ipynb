```python
# cell 1 - imports
from src.data_utils import load_raw, normalize_cols, parse_price_indian, parse_area_to_sqft, save_processed
import pandas as pd

# cell 2 - load & normalize
df = load_raw("house_prices.csv")
df = normalize_cols(df)
print("Columns after normalize:", df.columns.tolist())

# cell 3 - parse prices & areas
if "amountin_rupees" in df.columns:
    df["amount_in_rupees"] = df["amountin_rupees"].apply(parse_price_indian)
if "price_in_rupees" in df.columns:
    df["price_in_rupees"] = df["price_in_rupees"].apply(parse_price_indian)

for col in ["carpet_area", "super_area", "plot_area", "dimensions"]:
    if col in df.columns:
        df[col + "_sqft"] = df[col].apply(parse_area_to_sqft)

# create has_plot_area flag
if "plot_area_sqft" in df.columns:
    df["has_plot_area"] = df["plot_area_sqft"].notna()
else:
    df["has_plot_area"] = False

# quick sanity
print("Sample parsed amounts:", df["amount_in_rupees"].dropna().head().tolist())
print("Sample carpet_area_sqft:", df["carpet_area_sqft"].dropna().head().tolist())

# save
save_processed(df, "real_estate_cleaned_v1.csv")
```
