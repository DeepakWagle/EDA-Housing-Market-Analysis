```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from src.model import train_baseline, evaluate, save_model
from src.data_utils import load_processed

df = load_processed("house_prices_feature_v2.csv")

target = "amount_in_rupees"
df = df[df[target].notna() & (df[target] > 0)].copy()

num_cols = [
    "carpet_area_sqft",
    "super_area_sqft",
    "bathroom",
    "balcony",
    "floor_num"
]

num_cols = [c for c in num_cols if c in df.columns]

for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")

df[num_cols] = df[num_cols].fillna(df[num_cols].median())

assert df[num_cols].isna().sum().sum() == 0, "NaNs still exist in X!"

X = df[num_cols]
y = df[target]

scaler = StandardScaler()
X_scaled = pd.DataFrame(
    scaler.fit_transform(X),
    columns=num_cols
)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

model = train_baseline(X_train, y_train)
print("Baseline evaluation:", evaluate(model, X_test, y_test))
save_model(model, "../models/baseline_linear_amount.joblib")

```
